```python
def myinfo(x):
    if(x['dismissal_kind']==0):
        return 'not out'
    elif(x['dismissal_kind']=='caught'):
        return 'c '+ x['fielder'] +' b ' +x['bowler']
    elif(x['dismissal_kind']=='bowled'):
        return 'b '+ x['bowler']
    elif(x['dismissal_kind']=='lbw'):
        return 'lbw b '+ x['bowler']
    elif((x['dismissal_kind']=='run out') & (x['fielder']==0)):
        return 'run out ' + x['bowler']
    elif(x['dismissal_kind']=='run out'):
        return 'run out ' + str(x['fielder'])
    elif(x['dismissal_kind']=='retired hurt'):
        return 'retired hurt'
    elif(x['dismissal_kind']=='stumped'):
        return 'stumped ' + x['fielder'] +' b '+x['bowler']
    elif(x['dismissal_kind']=='caught and bowled'):
        return 'c ' +'&b '+ x['bowler']
    elif(x['dismissal_kind']=='hit wicket'):
        return 'hit wicket ' + ' b '+x['bowler']
    else:
        return 'obstructing the field'
```

```python
scorecard['info']=scorecard.apply(myinfo,axis=1)
```
